```python
import matplotlib.pyplot as plt
```

```python

countries_to_plot = ['United States', 'Canada', 'Brazil', 'Japan', 'Germany', 'France' ,'India']


filtered_data = df[df['Country Name'].isin(countries_to_plot)]


plt.bar (filtered_data['Country Name'], filtered_data['2022'])
plt.xlabel ('Country Name')
plt.ylabel ('Population in 2022')
plt.title('Population of selected countries in 2022')
plt.xticks(rotation=45, ha='right') 
plt.show()
```

```python


df_long = pd.melt(df, id_vars=['Country Name', 'Country Code'], 
                  var_name='Year', value_name='Population')


df_long['Year'] = pd.to_numeric(df_long['Year'])


def plot_country_data(country_name):
    
    country_data = df_long[df_long['Country Name'] == country_name]
    
    
    plt.figure(figsize=(10, 6))
    plt.plot(country_data['Year'], country_data['Population'], marker='o')
    plt.title(f'{country_name} Population Over Time')
    plt.xlabel('Year')
    plt.ylabel('Population')
    plt.grid(True)
    plt.show()


plot_country_data('United States')


plot_country_data('China')
plot_country_data('India')
```
